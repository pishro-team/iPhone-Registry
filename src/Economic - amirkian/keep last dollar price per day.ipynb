Read dollar dataset

```python
import pandas as pd

file_path = 'dollar.xlsx'
df = pd.read_excel(file_path)
df.head()
```

Combine Field Date with Field Time and sort by DateTime

```python
df['DateTime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str))
df = df.sort_values(by='DateTime')
df.head()
```

Group by day and keep only the latest price per day

```python
df['Date'] = df['DateTime'].dt.date
df = df.groupby('Date').last().reset_index()
df = df.sort_values(by='Date', ascending=False)
df.head()
```

```python
df.to_excel("dollar.xlsx", index=False)
```
